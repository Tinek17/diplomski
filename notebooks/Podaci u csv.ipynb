```python
pollutants = ['D_PM10K_mean','N_PM10K_mean','O_PM10K_mean','S_PM10K_mean', 'W_PM10K_mean']
```

```python
traffic_data = pd.read_csv('../data/promet.csv',index_col=0)
```

```python
satelit_data = pd.read_csv('../data/satelit_data.csv',index_col=0)
```

```python
temporal = ['year', 'dayofyear', 'month_Apr', 'month_Aug', 'month_Dec', 'month_Feb',
       'month_Jan', 'month_Jul', 'month_Jun', 'month_Mar', 'month_May',
       'month_Nov', 'month_Oct', 'month_Sep', 'weekday_Friday',
       'weekday_Monday', 'weekday_Saturday', 'weekday_Sunday',
       'weekday_Thursday', 'weekday_Tuesday', 'weekday_Wednesday',
       'season_fall', 'season_spring', 'season_summer', 'season_winter',
       'holiday', 'holiday_school']
```

```python
sites = ['DonBosco','Nord','Ost','Sud','West']

satelit = ['Temperature_Air_2m_Mean_24h', 'Cloud_Cover_Mean',
       'Wind_Speed_10m_Mean', 'Dew_Point_Temperature_2m_Mean',
       'Vapour_Pressure_Mean']
```

```python
def split_data(var_name, site_name, PM, rest_data):
    temp_y = PM[var_name]
    temp_y.name = var_name
    temp_temporal = rest_data[temporal]
    temp_x = rest_data.filter(like=site_name, axis=1)
    temp_x = pd.concat([temp_x, temp_temporal,temp_y], axis=1)
    return temp_x
```

```python
def prep_data_satelite(temp_x, site, data):
    weather_data = temp_x.filter(like=site, axis=1)
    temp_x = temp_x.drop(columns=weather_data.columns, axis=1)
    temp_x = pd.concat([temp_x, data], axis=1)
    return temp_x
```

```python
def traffic_for_station(site_temp, data_temp):
    traffic_temp = data_temp.filter(like=site_temp, axis=1)
    return traffic_temp
```

```python
for varname,site in zip(pollutants,sites):
    df = split_data(varname,site,PM,rest_data)
    #df.to_csv(f'../data_for_testing/local_{varname}_{site}.csv')
    df_1 = prep_data_satelite(df,site,satelit_data)
    #df_1.to_csv(f'../data_for_testing/satelite_{varname}_{site}.csv')
    df_2 = pd.concat([df,satelit_data],axis=1)
    #df_2.to_csv(f'../data_for_testing/all_meteo_{varname}_{site}.csv')
    traffic_station = traffic_for_station(site,traffic_data)
    df_3 = pd.concat([df,traffic_station],axis=1)
    #df_3.to_csv(f'../data_for_testing/local_traffic_{varname}_{site}.csv')
    df_4 = pd.concat([df_1,traffic_station],axis=1)
    #df_4.to_csv(f'../data_for_testing/satelit_traffic_{varname}_{site}.csv')
    df_5 = pd.concat([df_2,traffic_station],axis=1)
    #df_5.to_csv(f'../data_for_testing/all_traffic_{varname}_{site}.csv')
```
