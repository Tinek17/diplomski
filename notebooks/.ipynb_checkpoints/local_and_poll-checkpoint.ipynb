```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
PM = pd.read_csv(r'../data/PM.csv')
```

```python
local = pd.read_csv(r'../data/local_weather.csv')
```

```python
PM = PM.rename(columns={'Unnamed: 0':'index'})
```

```python
PM
```

```python
PM = PM.set_index('index')
```

```python
PM
```

```python
PM.reset_index(inplace=True)
```

```python
PM['index'] = pd.to_datetime(PM['index'])
```

```python
PM = PM.groupby(pd.Grouper(key='index', freq='D')).mean()
```

```python
PM
```

```python
PM = PM['2018-01-01':'2020-03-15']
```

```python
sites = ['Don Bosco','Sjever','Istok','Jug','Zapad',]
```

```python
PM.index = pd.to_datetime(PM.index)
```

```python
PM.index
```

```python
def pollutanti(site_1,pollution):
    df_fun = PM.copy()
    Year = df_fun.index
    x = df_fun[pollution]
    plt.figure(figsize=(20,8))
    plt.plot(Year, x)
    plt.title(site_1)
    plt.xlabel('Date')
    plt.ylabel('Lebdeće čestice')
    plt.show()
    
```

```python
for site,i in zip(sites,PM.columns):
    graf = pollutanti(site,i)
```

```python
local = local.rename(columns={'Unnamed: 0':'index'})
```

```python
local = local.set_index('index')
```

```python
local.reset_index(inplace=True)
```

```python
local['index'] = pd.to_datetime(local['index'])
```

```python
local = local.groupby(pd.Grouper(key='index', freq='D')).mean()
```

```python
local = local['2018-01-01':'2020-03-15']
```

```python
def weather_fun(weather):
    df_fun = local.copy()
    Year = df_fun.index
    x = df_fun[weather]
    plt.figure(figsize=(20,8))
    plt.plot(Year, x)
    plt.title(weather)
    plt.xlabel('Date')
    plt.ylabel(weather)
    plt.show()
    
```

```python
for i in local.columns:
    graf = weather_fun(i)
```
